```python
from __future__ import print_function
import pandas as pd
import shutil
import os
import sys
```

```python
df=pd.read_excel('ronin-airwire-dataset.xlsx')
```

```python
labels=df['Class']  ## Labels - Sub Folder Names
```

```python
labels
```

```python
names=df['Picture No'] ##Image Names
```

```python
names
```

```python
train_dir='dataset'  ## Source Directory
split=r'C:\\Users\\ASUS\\AirWire_Split_Data\\' ## Destination Directory
```

```python
for i in range(len(names)):
    if not os.path.exists(split + str(labels[i])):  ## If sub-folder doesn't exist
        os.mkdir(split+str(labels[i])) ## then, create the folder
    src = train_dir +'/'+str(names[i])+'.JPG'  ## source image name
    dest= split + str(labels[i])+'/'+str(names[i])+'.JPG'  ## new copied image name
    try:
        shutil.copy(src,dest) ## .copy to copy, .move to move
        print('Successful')  
    except IOError as e:
        print("Unable to copy")
    except:
        print('Unexpected error: {}'.format(sys.exc_info()))
        
```
